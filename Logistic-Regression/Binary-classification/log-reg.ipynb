```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_classes=2,n_features=10,n_clusters_per_class=1,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
# Model training
from sklearn.linear_model import LogisticRegression
```

```python
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
logistic.predict_proba(X_test)
```

```python
# performance metrics
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
acc_score=accuracy_score(y_true=y_test,y_pred=y_pred)
acc_score
```

```python
conf_mat=confusion_matrix(y_true=y_test,y_pred=y_pred)
conf_mat
```

```python
print(classification_report(y_true=y_test,y_pred=y_pred))
```

# Hyperparameter tuning and Cross Validation

```python
logistic=LogisticRegression()

penalty=["l2"]
c_vals=[100,10,1.0,0.1,0.01]
solver=["lbfgs","liblinear","sag","saga"]
```

```python
params=dict(penalty=penalty,C=c_vals,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold

cv=StratifiedKFold(n_splits=5)
```

```python
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(estimator=logistic,param_grid=params,scoring="accuracy",cv=cv,n_jobs=-1,error_score="raise")
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
print(y_pred)
```

# Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV

logistic=LogisticRegression()

random=RandomizedSearchCV(estimator=logistic,param_distributions=params,scoring="accuracy",cv=5)
```

```python
random.fit(X_train,y_train)
```

```python
random.best_score_
```

```python
random.best_params_
```

```python
y_pred=random.predict(X_test)
y_pred
```
