```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
X,y=make_classification(n_samples=1000,n_classes=2,n_clusters_per_class=1,random_state=42)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# generate a dummy model
dummy_model_prob=[0 for _ in range(len(y_test))]
dummy_model_prob
```

```python
# train our actual model
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
model_prob=model.predict_proba(X_test)
model_prob
```

```python
model_prob=model_prob[:,1]
model_prob
```

```python
dummy_model_auc=roc_auc_score(y_true=y_test,y_score=dummy_model_prob)
model_auc=roc_auc_score(y_true=y_test,y_score=model_prob)
print(dummy_model_auc)
print(model_auc)
```
