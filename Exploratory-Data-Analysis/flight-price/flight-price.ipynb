#### Let's handle "Date_of_Journey" column

```python

flight_df["Date"]=flight_df["Date_of_Journey"].str.split("/").str[0]
flight_df["Month"]=flight_df["Date_of_Journey"].str.split("/").str[1]
flight_df["Year"]=flight_df["Date_of_Journey"].str.split("/").str[2]
```

```python
flight_df.head()
```

```python
flight_df.info()
```

```python
# convert them into int type
flight_df["Date"]=flight_df["Date"].astype(int)
flight_df["Month"]=flight_df["Month"].astype(int)
flight_df["Year"]=flight_df["Year"].astype(int)
```

```python
flight_df.info()
```

```python
# now drop the date_of_journey column
flight_df.drop(["Date_of_Journey"],axis=1,inplace=True)
```

```python
flight_df.head()
```

#### Let's handle "Arrival_Time" column

```python
flight_df["Arrival_Time"]=flight_df["Arrival_Time"].apply(lambda x:x.split(" ")[0])
```

```python
flight_df.head()
```

```python
flight_df["Arrival_Hours"]=flight_df["Arrival_Time"].str.split(":").str[0]
flight_df["Arrival_Minutes"]=flight_df["Arrival_Time"].str.split(":").str[1]
```

```python
flight_df.head()
```

```python
flight_df.info()
```

```python
flight_df["Arrival_Hours"]=flight_df["Arrival_Hours"].astype(int)
flight_df["Arrival_Minutes"]=flight_df["Arrival_Minutes"].astype(int)
```

```python
flight_df.info()
```

```python
flight_df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
flight_df.head()
```

#### Let's handle 'Dep_Time' column

```python
flight_df["Departure_Hours"]=flight_df["Dep_Time"].str.split(":").str[0]
flight_df["Departure_Minutes"]=flight_df["Dep_Time"].str.split(":").str[1]
```

```python
flight_df.head()
```

```python
flight_df["Departure_Hours"]=flight_df["Departure_Hours"].astype(int)
flight_df["Departure_Minutes"]=flight_df["Departure_Minutes"].astype(int)
```

```python
flight_df.info()
```

```python
flight_df.drop("Dep_Time",inplace=True,axis=1)
```

```python
flight_df.head(2)
```

#### let's handle "Duration"

```python
flight_df["Duration_Hours"]=flight_df["Duration"].str.split(" ").str.split("h").str[0]
flight_df["Duration_Minutes"]=flight_df["Duration"].str.split(" ").str[1].str.split("m").str[0]
```

```python
flight_df.head()
```

```python
flight_df["Duration_Minutes"].fillna(0,inplace=True)
```

```python
# flight_df["Duration_Hours"]=flight_df["Duration_Hours"].astype(int)
flight_df["Duration_Minutes"]=flight_df["Duration_Minutes"].astype(int)
```

```python
flight_df.info()
```

```python
flight_df.head()
```

```python
flight_df.drop("Duration",axis=1,inplace=True)
```

```python
flight_df.head()
```

#### Perform OHE on the Airline column

```python
flight_df["Airline"].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
airline_encode=OneHotEncoder()
```

```python
encoded_airlines=airline_encode.fit_transform(flight_df[["Airline"]]).toarray()
```

```python
encoded_airlines_df=pd.DataFrame(encoded_airlines,columns=airline_encode.get_feature_names_out())
encoded_airlines_df
```

```python
flight_df.tail()
```

```python
flight_df.reset_index(drop=True,inplace=True)
```

```python
flight_df=pd.concat([flight_df,encoded_airlines_df],axis=1)
flight_df
```
