```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.DataFrame({
    "Object":["object1","object2","object3","object4","object5"],
    "Color":["red","green","blue","green","red"]
})
```

```python
df
```

```python
encoder=OneHotEncoder()
```

```python
spars_mat=encoder.fit_transform(df[["Color"]])
```

```python
encoded_arr=spars_mat.toarray()
print(encoded_arr)
```

```python
encoded_colors=pd.DataFrame(encoded_arr,columns=encoder.get_feature_names_out())
```

```python
encoded_colors
```

```python
encoded_df=pd.concat([df["Object"],encoded_colors],axis=1)
encoded_df
```

```python
import seaborn as sns
```

```python
tips_df=sns.load_dataset('tips')
```

```python
tips_df
```

```python
sex_encoder=OneHotEncoder()
smoker_encoder=OneHotEncoder()
time_encoder=OneHotEncoder()
```

```python
encoded_sex=sex_encoder.fit_transform(tips_df[["sex"]]).toarray()
```

```python
sex_encoded_df=pd.DataFrame(encoded_sex,columns=sex_encoder.get_feature_names_out())
sex_encoded_df
```

```python
tips_encoded=tips_df.drop(columns="sex")
tips_encoded=pd.concat([tips_encoded,sex_encoded_df],axis=1)
tips_encoded
```

```python
encoded_smoker=smoker_encoder.fit_transform(tips_encoded[["smoker"]]).toarray()
```

```python
smoke_encoded_df=pd.DataFrame(encoded_smoker,columns=smoker_encoder.get_feature_names_out())
smoke_encoded_df
```

```python
tips_encoded=tips_df.drop(columns="smoker")
tips_encoded=pd.concat([tips_encoded,smoke_encoded_df,sex_encoded_df],axis=1)
tips_encoded
```

```python
encoded_time=time_encoder.fit_transform(tips_encoded[["time"]]).toarray()
```

```python
encoded_time_df=pd.DataFrame(encoded_time,columns=time_encoder.get_feature_names_out())
encoded_time_df
```

```python
tips_encoded=tips_df.drop(columns=["sex","smoker","time"])
tips_encoded=pd.concat([tips_encoded,smoke_encoded_df,sex_encoded_df,encoded_time_df],axis=1)
tips_encoded
```
