```python
import numpy as np
```

```python
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,67,74]
```

```python
mini,Q1,median,Q3,maxi=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
print(mini)
print(Q1)
print(median)
print(Q3)
print(maxi)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*(IQR)
higher_fence=Q3+1.5*(IQR)
```

```python
print(lower_fence)
print(higher_fence)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(lst_marks) # no outliers in the boxplot
plt.show()
```
