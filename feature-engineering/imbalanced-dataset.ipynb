### 1. Up Sampling

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(42)


n_samples=1000
class_0_ratio=0.9
n_class_0=int(class_0_ratio*n_samples)
n_class_1=n_samples-n_class_0
```

```python
n_class_0,n_class_1
```

```python
class_0=pd.DataFrame({
    "feature_1":np.random.normal(loc=0,scale=1,size=n_class_0),
    "feature_2":np.random.normal(loc=0,scale=1,size=n_class_0),
    "target":[0]*n_class_0
})
```

```python
class_0
```

```python
class_1=pd.DataFrame({
    "feature_1":np.random.normal(loc=2,scale=1,size=n_class_1),
    "feature_2":np.random.normal(loc=2,scale=1,size=n_class_1),
    "target":[1]*n_class_1
})
```

```python
class_1
```

```python
imbalanced_df=pd.concat([class_0,class_1]).reset_index(drop=True)
imbalanced_df
```

```python
imbalanced_df["target"].value_counts()
```

```python
minority_df=imbalanced_df[imbalanced_df["target"]==1]
minority_df
```

```python
majority_df=imbalanced_df[imbalanced_df["target"]==0]
majority_df
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsampled=resample(minority_df,replace=True,n_samples=len(majority_df),random_state=42)
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled=pd.concat([majority_df,df_minority_upsampled])
df_upsampled
```

```python
df_upsampled["target"].value_counts()
```

### 2.Down Sampling

```python
np.random.seed(42)


n_samples=1000
class_0_ratio=0.9
n_class_0=int(class_0_ratio*n_samples)
n_class_1=n_samples-n_class_0
```

```python
class_0=pd.DataFrame({
    "feature_1":np.random.normal(loc=0,scale=1,size=n_class_0),
    "feature_2":np.random.normal(loc=0,scale=1,size=n_class_0),
    "target":[0]*n_class_0
})


class_1=pd.DataFrame({
    "feature_1":np.random.normal(loc=2,scale=1,size=n_class_1),
    "feature_2":np.random.normal(loc=2,scale=1,size=n_class_1),
    "target":[1]*n_class_1
})
```

```python
imbalanced_df=pd.concat([class_0,class_1])
imbalanced_df["target"].value_counts()
```

```python
df_majority=imbalanced_df[imbalanced_df["target"]==0]
df_minority=imbalanced_df[imbalanced_df["target"]==1]
```

```python
df_majority_downsampled=resample(
    df_majority,
    replace=False,
    n_samples=len(df_minority),
    random_state=42
)
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled=pd.concat([df_majority_downsampled,df_minority])
df_downsampled
```

```python
df_downsampled["target"].value_counts()
```
